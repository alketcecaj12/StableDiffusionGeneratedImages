```python
import time
import keras_cv
from tensorflow import keras
import matplotlib.pyplot as plt
import functions
import matplotlib.pyplot as plt
```

```python
model = keras_cv.models.StableDiffusion(img_width=512, img_height=512, jit_compile=False)
```

```python
prompt = """A national geographic photo of a chinese cormorant fisherman, a solitary lantern at the bow of his bamboo raft casting a warm glow on his nets, blue hour, in a karst landscape in yangshuo china, underexposed"""
```

```python
images = model.text_to_image(prompt, batch_size=2)
```

```python
functions.plot_images(images)
```

```python
# Save the image using matplotlib.pyplot.imsave()
plt.imsave('img/NationalGeographic.png', images[1], cmap='gray')
```
